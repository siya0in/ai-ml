```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv("final_3.5.csv")
df2=pd.read_csv("final_4.5.csv")
df3=pd.read_csv("final_5.5.csv")
df4=pd.read_csv("final_6.5.csv")
df5=pd.read_csv("final_7.5.csv")
```

```python
print(df1)
```

```python
datasets=[df1,df2,df3,df4,df5]
```

```python
df=pd.concat(datasets)
```

```python
print(df)
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df['time'] = df['time'].dt.time
```

```python
print(df)
```

```python
print(df)
```

```python
df['timestamp'] = pd.to_datetime(df['date'].astype(str) + ' ' + df['time'].astype(str))
```

```python
print(df)
```

```python
fig, ax1 = plt.subplots(figsize=(20, 8))

df['average insect count']=df['mean / 10 min']*20

# Plot the first set of data (Temp Mean, Humidity Mean, Average Insect Count)
ax1.plot(df['timestamp'], df['temp_mean'], color='red', label='Temp Mean')
ax1.plot(df['timestamp'], df['humidity_mean'], color='blue', label='Humidity Mean')
ax1.plot(df['timestamp'], df['mean / 10 min'], color='green', label='Average Insect Count')

# Set labels and title for the first y-axis
ax1.set_xlabel('Time')
ax1.set_ylabel('Temp Mean / Humidity Mean / Average Insect Count')

# Create a second y-axis
ax2 = ax1.twinx()

# Plot the second set of data (Light Mean)
ax2.plot(df['timestamp'], df['light_mean'], color='orange', label='Light Mean')

# Set label for the second y-axis
ax2.set_ylabel('Light Mean')
ax1.set_xticks(df['timestamp'][::10])

# Display the legend
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```
